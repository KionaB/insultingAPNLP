```python
#imports
import re
from operator import contains
```

```python
#prompt input
first_insult = input("Insult me, I dare you [template: [X] are/is as [Y] as a [Z]]")
print("Your insult:", first_insult)
```

```python
#regex based filter for template:  [X] are/is as [Y] as a [Z], works for multiple words per var
template_first= r"( are as | is as )"
first_compiled=re.compile(template_first)
pt1 = re.sub(first_compiled,"|",first_insult)
template_second= (r"( as a | as an )")
second_compiled=re.compile(template_second)
variables = re.sub(second_compiled,"|",pt1).split("|")
print(variables)
subject=variables[0]
insult_scale=variables[1]
comparator=variables[2]
```

```python
#Generate worse insult on given scale
worse_comparator=" "
```

```python
#Output
comeback = "Comeback: "+subject
if " is as " in first_insult:
    comeback=comeback+" is as "+insult_scale
else:
    comeback=comeback+" are as "+insult_scale
if worse_comparator[0] in ['a','e','i','o','u']:
    comeback=comeback+" as an "+worse_comparator
else:
    comeback=comeback+" as a "+worse_comparator

print(comeback)
```
